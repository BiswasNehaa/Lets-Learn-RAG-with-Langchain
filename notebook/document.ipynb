```python
### Text splitting get into chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter

def split_documents(documents,chunk_size=1000,chunk_overlap=200):
    """Split documents into smaller chunks for better RAG performance"""
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", " ", ""]
    )
    split_docs = text_splitter.split_documents(documents)
    print(f"Split {len(documents)} documents into {len(split_docs)} chunks")
    
    # Show example of a chunk
    if split_docs:
        print(f"\nExample chunk:")
        print(f"Content: {split_docs[0].page_content[:200]}...")
        print(f"Metadata: {split_docs[0].metadata}")
    
    return split_docs
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
import uuid
from typing import List,Dict,Any,Tuple
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from typing import List

class EmbeddingManager:
    """Handles doc embedding generation using sentence transformers"""
    
    def __init__(self, model_name: str = "all-MiniLM-L6-v2"):
        """
        Initialize the Embedding Manager
        """
        self.model_name = model_name
        self.model = None
        self._load_model()
    
    def _load_model(self):
        """Load the sentence transformer model"""
        try:
            print(f"Loading embedding model: {self.model_name}")
            self.model = SentenceTransformer(self.model_name)
            print(f"Model loaded successfully. Dimension: {self.model.get_sentence_embedding_dimension()}")
        except Exception as e:
            print(f"Error loading model {self.model_name}: {e}")
            raise
    
    def generate_embeddings(self, texts: List[str]) -> np.ndarray:
        """Generate embeddings for a list of texts"""
        if self.model is None:
            raise ValueError("Model not loaded")
        
        print(f"Generating embeddings for {len(texts)} texts")
        embeddings = self.model.encode(texts, show_progress_bar=True)
        print(f"Generated embeddings shape: {embeddings.shape}")
        return embeddings


# Initialize
embedding_manager = EmbeddingManager()
```

Vector Store

```python
# Import required libraries

import os                      # Used to create folders on your system
import uuid                    # Used to generate unique IDs for each document
import chromadb                # ChromaDB is our vector database
import numpy as np             # Used for handling embeddings (arrays of numbers)
from typing import List, Any   # Helps define input types (optional but good practice)


class VectorStore:
    """
    This class manages storing and retrieving embeddings in ChromaDB.
    Think of it like a smart storage system for your document chunks.
    """

    def __init__(self, collection_name: str = "pdf_documents", persist_directory: str = "../data/vector_store"):
        """
        This runs automatically when you create the object.

        Args:
            collection_name: Name of your database collection
            persist_directory: Folder where data will be saved
        """

        # Store the collection name
        self.collection_name = collection_name

        # Store the folder path where DB will be saved
        self.persist_directory = persist_directory

        # This will hold the database client (connection)
        self.client = None

        # This will hold the actual collection (like a table)
        self.collection = None

        # Initialize the database
        self._initialize_store()


    def _initialize_store(self):
        """
        This function:
        1. Creates folder (if not exists)
        2. Connects to ChromaDB
        3. Creates or loads a collection
        """

        try:
            # Create directory if it does not exist
            os.makedirs(self.persist_directory, exist_ok=True)

            # Create ChromaDB client (this connects to DB)
            self.client = chromadb.PersistentClient(path=self.persist_directory)

            # Get existing collection OR create new one
            self.collection = self.client.get_or_create_collection(
                name=self.collection_name,
                metadata={"description": "PDF document embeddings for RAG"}
            )

            # Print info
            print(f"✅ Vector store initialized")
            print(f"Collection name: {self.collection_name}")
            print(f"Existing documents: {self.collection.count()}")

        except Exception as e:
            print(f"❌ Error initializing vector store: {e}")
            raise


    def add_documents(self, documents: List[Any], embeddings: np.ndarray):
        """
        This function stores your chunks + embeddings into ChromaDB

        Args:
            documents: List of document chunks (LangChain Documents)
            embeddings: Numerical vectors for each document
        """

        # Check if number of documents matches number of embeddings
        if len(documents) != len(embeddings):
            raise ValueError("Number of documents must match number of embeddings")

        print(f"Adding {len(documents)} documents to vector store...")

        # Prepare empty lists to store data
        ids = []               # Unique IDs for each document
        metadatas = []        # Extra info about documents
        documents_text = []   # Actual text content
        embeddings_list = []  # Embeddings converted to list

        # Loop through each document and embedding
        for i, (doc, embedding) in enumerate(zip(documents, embeddings)):

            # Create a unique ID for each document
            doc_id = f"doc_{uuid.uuid4().hex[:8]}_{i}"

            # Add ID to list
            ids.append(doc_id)

            # Get metadata (if exists)
            metadata = dict(doc.metadata) if doc.metadata else {}

            # Add extra useful info
            metadata['doc_index'] = i
            metadata['content_length'] = len(doc.page_content)

            # Save metadata
            metadatas.append(metadata)

            # Save document text
            documents_text.append(doc.page_content)

            # Convert embedding (numpy → list) and store
            embeddings_list.append(embedding.tolist())

        # Now insert everything into ChromaDB
        try:
            self.collection.add(
                ids=ids,
                embeddings=embeddings_list,
                metadatas=metadatas,
                documents=documents_text
            )

            print(f"✅ Successfully added {len(documents)} documents")
            print(f"📊 Total documents in DB: {self.collection.count()}")

        except Exception as e:
            print(f"❌ Error adding documents: {e}")
            raise


# Create the vector store object (this will initialize DB)
vectorstore = VectorStore()
```

```python
# Step 1: Extract text
texts = [doc.page_content for doc in chunks]

# Step 2: Generate embeddings
embeddings = embedding_manager.generate_embeddings(texts)

# Step 3: Store in vector DB
vectorstore.add_documents(chunks, embeddings)
```

RAG Retriever pipeline from vector store

```python


class RAGRetriever:
    """
    This class is responsible for retrieving (searching) relevant documents
    from the vector database using a user query.
    """

    def __init__(self, vector_store, embedding_manager):
        """
        Constructor (runs when object is created)

        Args:
            vector_store → your ChromaDB wrapper (VectorStore)
            embedding_manager → converts text into embeddings (numbers)
        """

        # Save reference to vector database
        self.vector_store = vector_store

        # Save embedding generator
        self.embedding_manager = embedding_manager


    def retrieve(self, query: str, top_k: int = 5, score_threshold: float = 0.0) -> List[Dict[str, Any]]:
        """
        Retrieve relevant documents for a query

        Args:
            query → user question
            top_k → how many results to return
            score_threshold → minimum similarity score (filter weak results)

        Returns:
            List of dictionaries (each contains document + metadata + score)
        """

        # Print debugging info
        print(f"\n🔍 Retrieving documents for query: '{query}'")
        print(f"Top K: {top_k}, Score threshold: {score_threshold}")

        # =========================
        # STEP 1: Convert query → embedding
        # =========================
        # Convert query text into numbers (embedding)
        query_embedding = self.embedding_manager.generate_embeddings([query])[0]


        # =========================
        # STEP 2: Search in vector DB
        # =========================
        try:
            results = self.vector_store.collection.query(
                query_embeddings=[query_embedding.tolist()],  # Convert to list for ChromaDB
                n_results=top_k
            )

            # =========================
            # STEP 3: Prepare results container
            # =========================
            retrieved_docs = []

            # Check if results exist
            if results['documents'] and results['documents'][0]:

                # Extract returned data
                documents = results['documents'][0]
                metadatas = results['metadatas'][0]
                distances = results['distances'][0]
                ids = results['ids'][0]

                # =========================
                # STEP 4: Process each result
                # =========================
                for i, (doc_id, document, metadata, distance) in enumerate(
                        zip(ids, documents, metadatas, distances)):

                    # Convert distance → similarity score
                    # (ChromaDB gives distance, we convert it)
                    similarity_score = 1 - distance

                    # Apply filtering (remove weak results)
                    if similarity_score >= score_threshold:

                        # Store everything in dictionary
                        retrieved_docs.append({
                            'id': doc_id,                     # unique document ID
                            'content': document,              # actual text chunk
                            'metadata': metadata,             # extra info
                            'similarity_score': similarity_score,  # how similar
                            'distance': distance,             # raw distance
                            'rank': i + 1                     # position (1st, 2nd...)
                        })

                print(f"✅ Retrieved {len(retrieved_docs)} documents (after filtering)")

            else:
                print("❌ No documents found")

            return retrieved_docs


        # =========================
        # ERROR HANDLING
        # =========================
        except Exception as e:
            print(f"❌ Error during retrieval: {e}")
            return []
        
rag_retriver=RAGRetriever(vectorstore,embedding_manager)
```
