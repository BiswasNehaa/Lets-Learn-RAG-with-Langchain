```python
### Langchain + Typesense + Groq LLM + RAG Application

from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import Typesense
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_groq import ChatGroq
```

```python
import os
os.environ["GROQ_API_KEY"] = "Your Actaual api key"
```

```python
loader=TextLoader("text.txt")
documents=loader.load()   ### Data Ingestion
text_splitter=CharacterTextSplitter(chunk_size=1000, chunk_overlap=100)  ### Pre-processing
docs=text_splitter.split_documents(documents)   ### Text Splitting

embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
### using Typesense along with Langchain

docsearch=Typesense.from_documents(
    docs, 
    embeddings, 
    typesense_client_params={
        "host": "8hgopwksfm2d9lvcp-1.a2.typesense.net",
        "port": 443,
        "protocol": "https",
        "typesense_api_key": "Typesense master api key",
        "typesense_collection_name": "lang-chain"
    },
)
```

```python
query="What is RAG?"
find_docs=docsearch.similarity_search(query)
print(find_docs)
```

```python
### Retriver
retriever=docsearch.as_retriever()
retriever
```
